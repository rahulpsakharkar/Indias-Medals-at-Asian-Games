# India's Medals at Asian Games

```python
# Link to dataset
# https://indianathletics.in/medal-winners-of-asian-games/
```

```python
# imports

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Get data from website source

df = pd.read_html('https://indianathletics.in/medal-winners-of-asian-games/')[0]
df.head(2)
```

```python
df.set_index('S.NO', inplace = True)
df.head(2)
```

```python
# Bar graph showing medals own by India by type

plt.figure(figsize = (10,6))
df['RANK/PLACE'].value_counts().plot(kind = 'bar', color = ['silver', 'brown', 'gold'])

plt.title('India\'s Medals at Asian Games by Medal Type', fontsize = 16)
plt.xlabel('Medal Type', fontsize = 14)
plt.ylabel('Medal Count', fontsize = 14)
plt.xticks(rotation = 'horizontal', fontsize = 12)

plt.ylim([0,100]);
```

```python
# Calculate medal count in each type

medals = pd.crosstab(df['YEAR'], df['RANK/PLACE'])
medals.head(2)
```

```python
# Create a new column giving addition of all medals per year

medals['All'] = medals.sum(axis = 1)
medals.head(2)
```

```python
# Plot line graphs showing India's medals at Asian Games

plt.figure(figsize = (14,8))

plt.plot(medals.index, medals['Gold'], label = 'Gold', c = 'gold')
plt.plot(medals.index, medals['Silver'], label = 'Silver', c = 'silver')
plt.plot(medals.index, medals['Bronze'], label = 'Bronze', c = 'brown')
plt.plot(medals.index, medals['All'], label = 'All', c = 'red')

plt.legend(loc = 'best', fontsize = 14, frameon = False)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Number of Medals', fontsize = 14)
plt.title('India\'s Medals at Asian Games', fontsize = 17);
```
